## Data cleaning

```python
# suppression colonnes jugées inutiles

del df["backdrop_path"]

del df["homepage"]

del df["imdb_id"]

del df["keywords"]

del df["tagline"]

del df["spoken_languages"]

del df["poster_path"]
```

```python
# suppression des lignes ou "title" est null

df = df.dropna(subset="title")
```

```python
# suppression des lignes ou "runtime" est inférieur à 30

df = df[df['runtime'] >= 30]
```

```python
# suppression des lignes ou "budget" est inférieur à 10000

df = df[df['budget'] >= 10000]
```

```python
# modification du Dtype "release_date"

df["release_date"] = pd.to_datetime(df["release_date"], errors='coerce')
```

```python
# remplissage des valeurs nulles

df["release_date"] = df["release_date"].fillna(pd.to_datetime("1900-01-01"))

df["overview"] = df["overview"].fillna("Unknown")

df["genres"] = df["genres"].fillna("Unknown")

df["production_companies"] = df["production_companies"].fillna("Unknown")

df["production_countries"] = df["production_countries"].fillna("Unknown")
```

```python

#df = df[~df["title"].str.contains(r"- \\")]

#df = df[~df["original_title"].str.contains(r"- \\")]

#df["overview"] = df["overview"].str.replace(r"\\", "/", regex=True)
```

```python
# filtre sur les titres par les caractères

df = df[df["title"].str.contains(r"^[a-zA-Z0-9éèêç&@#?!\-_: ]+$")]
```

```python
# conservation des deux premiers genres uniquement

df['genres'] = df['genres'].apply(lambda x: ', '.join(x.split(', ')[0:2]))
```
